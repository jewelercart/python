```python
import matplotlib.pyplot as plt

# Scatter plot of sepal length vs. sepal width with color-coded species
plt.figure(figsize=(8, 6))
colors = {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}
for species, color in colors.items():
    subset = iris_data[iris_data['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species, c=color)

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.legend(loc='upper right', fontsize='small')
plt.show()
```

```python
# Bar plot of species counts
species_counts = iris_data['species'].value_counts()
species_counts.plot(kind='bar', color=['r', 'g', 'b'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Plot of Species Counts')
plt.xticks(rotation=0)
plt.show()
```

```python
# Line chart of petal length over the index
iris_data['petal_length'].plot()
plt.xlabel('Index')
plt.ylabel('Petal Length (cm)')
plt.title('Line Chart of Petal Length')
plt.show()
```

```python
# Histogram of petal width
plt.hist(iris_data['petal_width'], bins=20, color='purple', edgecolor='black')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()
```
